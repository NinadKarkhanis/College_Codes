```python
import pandas as pd
import numpy as np
import random as rd
import matplotlib.pyplot as plt
# Load the data
data = pd.read_csv('clustering.csv')
X = data[["LoanAmount", "ApplicantIncome"]]
# Number of clusters
K = 3

# Select random observations as initial centroids
Centroids = X.sample(n=K)

# Plot initial data and centroids
plt.scatter(X["ApplicantIncome"], X["LoanAmount"], c='black')
plt.scatter(Centroids["ApplicantIncome"], Centroids["LoanAmount"], c='red')
plt.xlabel('Annual Income')
plt.ylabel('Loan Amount (In Thousands)')
plt.show()

# Initialize variables
diff = 1
j = 0

while diff != 0:
    XD = X.copy()
    i = 1
    for index1, row_c in Centroids.iterrows():
        ED = []
        for index2, row_d in XD.iterrows():
            d1 = (row_c["ApplicantIncome"] - row_d["ApplicantIncome"])**2
            d2 = (row_c["LoanAmount"] - row_d["LoanAmount"])**2
            d = np.sqrt(d1 + d2)
            ED.append(d)
        X[i] = ED
        i = i + 1

    C = []
    for index, row in X.iterrows():
        min_dist = row[1]
        pos = 1
        for i in range(K):
            if row[i + 1] < min_dist:
                min_dist = row[i + 1]
                pos = i + 1
        C.append(pos)
    X["Cluster"] = C

    # Calculate new centroids
    Centroids_new = X.groupby(["Cluster"]).mean()[["LoanAmount", "ApplicantIncome"]]

    if j == 0:
        diff = 1
        j = j + 1
    else:
        diff = (
            (Centroids_new['LoanAmount'] - Centroids['LoanAmount']).sum()
            + (Centroids_new['ApplicantIncome'] - Centroids['ApplicantIncome']).sum()
        )
        Centroids = X.groupby(["Cluster"]).mean()[["LoanAmount", "ApplicantIncome"]]

# Define colors for plotting clusters
colors = ['blue', 'green', 'cyan']

# Plot the final clusters and centroids
for k in range(K):
    data = X[X["Cluster"] == k + 1]
    plt.scatter(data["ApplicantIncome"], data["LoanAmount"], c=colors[k])

plt.scatter(Centroids["ApplicantIncome"], Centroids["LoanAmount"], c='red')
plt.xlabel('Income')
plt.ylabel('Loan Amount (In Thousands)')
plt.show()
```
